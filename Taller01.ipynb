## Graficar Funciones

La función a graficar es:

$$ y(x) = x^3 - 7x^2 + 4x + 12a $$
- Tabla de Funciones

```python
#Definir la ecuacion
def eqn(n):
    return x ** 3 - 7 * x ** 2 + 4 * x + 12
```

```python
#Generación de puntos para la grafica
import numpy as np 
Xs, step = np.linspace(-10,10,num=150,retstep=True)
```

```python
Ys = []
for x in Xs:
    y = eqn(x)
    Ys.append(y)
    print(f"x = {x:.2f} \ty = {y:.2e}")
```

```python
from matplotlib import pyplot as plt

# import seaborn as sns ## alternativa chévere

plt.plot(Xs, Ys, marker=".", color="blue", linestyle=":")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Gráfica de función cúbica")
plt.grid()
plt.legend(["$y=x^3-7x^2+4x+12$"])

## TAREA hacer zoom

# Definir los límites para hacer zoom
plt.xlim(-5, 25)  # Ajusta estos valores para cambiar el zoom en el eje x
plt.ylim(-25, 40)  # Ajusta estos valores para cambiar el zoom en el eje y

plt.show()

```
