# Uso de librer√≠as Scipy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

x = np.linspace(-2, 4, 400)
# -2 l√≠mite inicial, 4 l√≠mite final incl., 400 n√∫mero de puntos

y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y
```

```python
def f_prime(x: float, *args):  # Agregamos *args para ignorar argumentos adicionales
    return 3 * x**2 - 6 * x + 1  # Derivada de f(x)

# Configuraci√≥n del m√©todo de Newton
x0 = 100
points = []  # Lista para almacenar los puntos evaluados

# Aplicaci√≥n del m√©todo de Newton con la derivada
x_r = newton(f, x0, fprime=f_prime, args=(points, True), maxiter=100)

print("Ra√≠z aproximada:", x_r)
print("Puntos evaluados:", points)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.scatter(x_r, f(x_r), color="red", zorder=5)  # Add the root to the plot
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
# Extract x values from points
x_points = [point[0] for point in points]

# Plot the x points
plt.figure(figsize=(8, 6))
plt.plot(x_points, "o-", label="x points")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of x points in points")
plt.xlabel("Index")
plt.ylabel("x value")
plt.legend()
plt.show()
```

```python
# Definimos la funci√≥n y su derivada
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y

def f_prime(x: float, *args):  # A√±adimos *args para ignorar los argumentos extra
    return 3 * x**2 - 6 * x + 1  # Derivada de f(x)

# Punto inicial
x0 = -1
points = []  # Lista para almacenar los puntos evaluados

# Aplicaci√≥n del m√©todo de Newton, especificando la derivada y obteniendo informaci√≥n completa
x_r, alg_info = newton(
    f,
    x0,
    fprime=f_prime,       # A√±adimos la derivada para usar el m√©todo de Newton
    args=(points, False),  # Argumentos adicionales para la funci√≥n `f`
    maxiter=50,
    disp=False,            # No mostrar mensajes de error cuando no converge
    full_output=True       # Devolver informaci√≥n adicional del algoritmo
)

print("Ra√≠z aproximada:", x_r)
print("Informaci√≥n del algoritmo:", alg_info)
print("Puntos evaluados:", points)
```
