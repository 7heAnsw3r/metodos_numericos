# Ejercicio Número 1 - Raices de una Funcion

Para la función:

f(x)=x3+2x2−5

encuentre la o las raíces.

```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 + 2*x**2 - 5

intervalos = [(i,i+1) for i in range(-100,100)]


raices = []

for a, b in intervalos:
    try:
        raiz = optimize.bisect(f, a, b, maxiter=100)
        raices.append(raiz)
        print(f"Raiz encontrada en el intervalo ({a},{b}): {raiz:.6f}")
    except ValueError:
        print(f"No se encontró raiz en el intervalo ({a},{b})")

x = np.linspace(-10,10,100)
y = f(x)

plt.figure(figsize=(10,5))
plt.plot(x,y, label="f(x)")
plt.scatter(raices, [0]*len(raices), color="red", label="Raices")
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Búsqueda de raíces con el método de bisección")
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(1, 2)
plt.ylim(min(y)*1.1, max(y)*1.1)  #

plt.show()
```

```python
# Ahora con el metodo de Newton

def f(x):
    return x**3 + 2*x**2 - 5

def df(x):
    return 3*x**2 + 4*x

def df2(x):
    return 6*x + 4

xs = np.arange(-10,10,10)

root = [optimize.newton(f, a )for a in xs]

x = np.linspace(-10,10,100)
y_vals = f(x)
plt.figure(figsize=(10,6))
plt.plot(x,y, label="f(x)")
plt.scatter(root, [0]*len(root), color="red", label="Raices")
plt.legend()
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Búsqueda de raíces con el método de Newton")
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(-10, 10)
plt.ylim(min(y_vals)*1.1, max(y_vals)*1.1)

print(root)
```

```python

# Utilizando las derivadas de las funciones para encontrar las raices de la funcion

xs = np.arange(-,5,100)

root2 = [optimize.newton(f, a, fprime=df, fprime2=df2) for a in xs]


x = np.linspace(-10,10,100)
y_vals = f(x)
plt.figure(figsize=(10,6))
plt.plot(x,y, label="f(x)")
plt.scatter(root, [0]*len(root), color="red", label="Raices")
plt.legend()
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Búsqueda de raíces con el método de Newton")
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(-10, 10)
plt.ylim(min(y_vals)*1.1, max(y_vals)*1.1)

print(root2)
```

# Resultados

Con el metodo de la biseccion solo encontramos las raices en el intervalo (1,2) y con el metodo de Newton encontramos las raices en los intervalos (-2.5,-2), (1,2) y (1,2) con las derivadas de la funcion.

Si utilizamos la derivada de la funcion debemos utilizar un intervalo de busqueda mas pequeño para encontrar las raices de la funcion, caso contrario apartir de la iteracion numero 50 no encontrara las raices de la funcion.
