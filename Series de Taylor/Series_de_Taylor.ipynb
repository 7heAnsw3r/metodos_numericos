# Aproximaciones de Taylor de cos(x) en el Intervalo [-6, 6]

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para generar el rango de valores para x
def generate_x_values(start=-6, end=6, num=400):
    return np.linspace(start, end, num)

# Funciones para calcular la aproximación de Taylor para cos(x) 
def taylor_approx_2(x):
    return 1 - (x**2) / 2

def taylor_approx_4(x):
    return taylor_approx_2(x) + (x**4) / 24


def taylor_approx_6(x):
    return taylor_approx_4(x) - (x**6) / 720

def plot_approximations(x, cos_x, approx_2, approx_4, approx_6):
    plt.figure(figsize=(10, 6))
    plt.plot(x, cos_x, label='cos(x)', color='red')
    plt.plot(x, approx_2, label='2 términos de Taylor', color='pink', linestyle='--')
    plt.plot(x, approx_4, label='4 términos de Taylor', color='purple', linestyle='--')
    plt.plot(x, approx_6, label='6 términos de Taylor', color='gray', linestyle='--')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Aproximaciones de Taylor de la función cos(x)")
    plt.legend()
    plt.grid(True)
    plt.xlim(-6, 6)
    plt.ylim(-1.5, 1.5)
    plt.show()

def main():
    x = generate_x_values()
    cos_x = np.cos(x)
    approx_2 = taylor_approx_2(x)
    approx_4 = taylor_approx_4(x)
    approx_6 = taylor_approx_6(x)
    plot_approximations(x, cos_x, approx_2, approx_4, approx_6)

# Ejecutar la función principal
if __name__ == "__main__":
    main()
```

# Función Asintota

```python
def generate_x_values(start=-1.25, end=1.25, num=400):
    return np.linspace(start, end, num)

# Función para calcular aproximaciones de Taylor para 1 / (1 - x)
def taylor_approx(x, terms):
    approximation = np.ones_like(x) # Este metodo crea un nuevo array con las mismas dimensiones y forma que el array x, esta es la base para crear las aproximaciones ya que nuestro array va a estar llenos de unos
    for n in range(1, terms):
        approximation += x**n
    return approximation

# Función para graficar la función original y las aproximaciones de Taylor
def plot_approximations(x, original_func, approximations):
    plt.figure(figsize=(10, 7))
    plt.plot(x, original_func, label=r'$\frac{1}{1 - x}$', color='blue')  # Función original
    colors = ['orange', 'brown', 'blue', 'purple', 'red', 'pink']
    for i, approx in enumerate(approximations, start=1):
        label = f'{i+1} términos: ' + ' + '.join([f'x^{j}' for j in range(i+1)])
        plt.plot(x, approx, label=label, color=colors[i-1], linestyle='--')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Aproximaciones de Taylor de 1 / (1 - x) con distintos términos")
    plt.legend()
    plt.grid(True)
    plt.ylim(-1, 10)
    plt.show()

def main():
    x = generate_x_values()
    original_func = 1 / (1 - x)
    approximations = [taylor_approx(x, i) for i in range(1, 6)]
    plot_approximations(x, original_func, approximations)


if __name__ == "__main__":
    main()
```

## Método ones_like()

```python
def taylor_approx(x, terms):
    approximation = np.ones_like(x)  # Base inicial llena de unos
    print(f"Inicial: {approximation}")

    for n in range(1, terms):
        term_value = x**n
        print(f"Termino {x}")
        print(f"Término {n}: {term_value}")
        approximation += term_value
        print(f"Aproximación después del término {n}: {approximation}")

    return approximation

# Ejemplo de uso
x = np.linspace(-1.25, 1.25, 5)
terms = 4
result = taylor_approx(x, terms)
print(f"Resultado final: {result}")
```

# Función Logaritmica

```python
import numpy as np
import matplotlib.pyplot as plt

# Función que calcula ln(x)
def f(x):
    return np.log(x)

# Función que calcula el polinomio de Taylor de grado 3 para ln(x)
def taylor_polynomial(x):
    return (x-1) - ((x-1)**2)/2 + ((x-1)**3)/3

# Función para generar los valores de x
def generate_x_values(start=0.1, end=3, num=400):
    return np.linspace(start, end, num)

def plot_functions(x, f_values, taylor_values):
    plt.figure(figsize=(10, 6))
    plt.plot(x, f_values, label='f(x) = ln(x)', color='blue')
    plt.plot(x, taylor_values, label='Polinomio de Taylor de grado 3', color='orange', linestyle='--')
    plt.title('Polinomio de Taylor de grado 3')
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    plt.ylim(-3, 2)  # Limitar el eje y para mejor visualización
    plt.xlim(0, 2)
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    x_values = generate_x_values()
    f_values = f(x_values)
    taylor_values = taylor_polynomial(x_values)
    plot_functions(x_values, f_values, taylor_values)


if __name__ == "__main__":
    main()
```
